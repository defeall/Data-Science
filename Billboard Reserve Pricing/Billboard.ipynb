# Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df2.drop('Price', axis=1)
Y = df2[['Price']]
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 1)
reg_model = LinearRegression()
reg_model.fit(x_train, y_train)
```

```python
intercept = reg_model.intercept_[0]
coef = reg_model.coef_[0]

print("The intercept is ",intercept)
print('-'*100)

for cf in zip(X.columns, coef):
   print('the coefficient for',cf[0],' is ',cf[1])
```

```python
Y = np.log(Y)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 1)
base_pred = np.mean(y_test)
base_pred
```

```python
from sklearn.metrics import mean_squared_error
base_pred = np.repeat(base_pred, len(y_test))
base_rmse = np.sqrt(mean_squared_error(y_test, base_pred))
base_rmse
```

```python
lgr = LinearRegression(fit_intercept = True)
model_lin = lgr.fit(x_train, y_train)
price_predictions = lgr.predict(x_test)
price_predictions[:5]
```

```python
lin_mse = mean_squared_error(y_test, price_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
r_lin_test = model_lin.score(x_test, y_test)
r_lin_train = model_lin.score(x_train, y_train)
print(r_lin_test, r_lin_train)
```

```python
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.ensemble import GradientBoostingRegressor

# define the model
model = GradientBoostingRegressor()
# define the evaluation procedure
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate the model
n_scores = cross_val_score(model, x_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# report performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
model.fit(x_train, y_train)
```

```python
r_lin_test = model.score(x_test, y_test)
r_lin_train = model.score(x_train, y_train)
print(r_lin_test, r_lin_train)
```
